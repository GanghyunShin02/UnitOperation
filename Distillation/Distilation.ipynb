```python
import pandas as pd
from CoolProp.CoolProp import PropsSI
import numpy as np

df=pd.read_excel(r"C:\Users\User\Desktop\단조실\3.단증류\distillation_frame.xlsx").ffill()

T=df['Temp'].iloc[0]
rho_alc=PropsSI('D','T',T,'P',101325,'Ethanol')
alp=PropsSI('P','T',T,'Q',1,'Ethanol')/101325

V_bottle=df['V_bottle'].dropna().iloc[0]


df['rho_w']=(df['w_bottle']-df['bottle'])/V_bottle
df['rho_s']=(df['s_bottle']-df['bottle'])/V_bottle

rho_w=df['rho_w'].iloc[0]

def concent(r):
    rr=1/r
    rrho_w=1/rho_w
    rrho_alc=1/rho_alc

    return (rr-rrho_w)/(rrho_alc-rrho_w)

def compos(p,m):
    M=(1-p)*m
    return (p*m/46)/((p*m/46)+(M/18))
    

df['concentration']=concent(df['rho_s'])

m_solvent=df['concentration']*(df['w_bottle']-df['bottle'])
m_water=(1-df['concentration'])*(df['w_bottle']-df['bottle'])

df['composition']=compos(df['concentration'],m_solvent)



```

```python

df['rho']=(df['bottled']-df['bottle'])/V_bottle
df['m_liq']=df['plasked']-df['plask']

df['concentration1']=concent(df['rho'])
df['m_solute']=df['concentration1']*df['m_liq']

df['composition1']=compos(df['concentration1'],df['m_liq'])







```

```python
X1=df.loc[df['type']=='solution','composition1'].iloc[0]
X2=df.loc[df['type']=='bottom','composition1'].iloc[0]

W1=df.loc[df['type']=='solution','m_liq'].iloc[0]
W2=df.loc[df['type']=='bottom','m_liq'].iloc[0]

f=(alp/(alp-1))*np.log((1-X2)/(1-X1))+(1/(alp-1))*np.log(X1/X2)

WW2=W1*np.exp(-f)
D=W1-WW2
beta=D/W1
xD=((W1*X1)-(WW2*X2))/D
'''
beta=1-np.exp(-f)
D=W1*beta
WW2=W1-D
xD=((W1*X1)-(WW2*X2))/D
'''


df['beta']=beta
df['D']=D
df['xD']=xD



df.to_excel(r"C:\Users\User\Desktop\단조실\3.단증류\distillation_result.xlsx")
```
